```python
import numpy as np
import pandas as pd

# Imputing null values in Insurance_Validity column with 'Not Available'
df['Insurance_Validity'] = df['Insurance_Validity'].fillna('Not Available')

# Imputing null values in Color column with 'Others'
df['Color'] = df['Color'].fillna('Others')

# Impute missing values in 'Kerb_Weight' using classification model
df_encoded['Kerb_Weight'] = df_encoded['Kerb_Weight'].replace('Kerb Weight', np.nan)
df_encoded = dc.impute_missing_values(df_encoded, target_column='Kerb_Weight', feature_columns=['Body_Type', 'Engine', 'Company', 'model', 'Fuel_Type', 'Transmission', 'centralVariantId'], 
    model_type='classification')

# Combine imputed 'Kerb_Weight' values from df_encoded into df
df['Kerb_Weight'] = df['Kerb_Weight'].replace('Kerb Weight', np.nan)
df['Kerb_Weight'] = df['Kerb_Weight'].combine_first(df_encoded['Kerb_Weight'])

# Ensure 'Registration_Year' is integer
current_year = pd.Timestamp.now().year
df['Registration_Year'] = df['Registration_Year'].astype(int)

# Add modelYear and Age_old_in_year columns based on Registration_Year
df['modelYear'] = df['modelYear'].astype(int)
df['Age_old_in_year'] = current_year - df['Registration_Year']

# Copy these columns to df_encoded for imputation purposes
df_encoded['Registration_Year'] = df['Registration_Year']
df_encoded['Age_old_in_year'] = df['Age_old_in_year']

# Handle Kms_Driven column, extracting numbers and handling missing data
df_encoded['Kms_Driven'] = df_encoded['Kms_Driven'].apply(
    lambda x: dc.extract_numbers(x, target_dtype=df['Kms_Driven'].dtype.type) if pd.notna(x) else x
)

# Fill Kms_Driven NaNs with -1, convert to int, and replace -1 with NaN
df_encoded['Kms_Driven'] = df_encoded['Kms_Driven'].fillna(-1).astype(int)
df_encoded['Kms_Driven'] = df_encoded['Kms_Driven'].replace(-1, np.nan)

# Impute missing values in 'Kms_Driven' using regression model
df_encoded = dc.impute_missing_values(df_encoded, target_column='Kms_Driven', feature_columns=['Registration_Year', 'Age_old_in_year'], 
    model_type='regression')

# Round Kms_Driven values and convert them to integers
df_encoded['Kms_Driven'] = df_encoded['Kms_Driven'].round(0).astype(int)

# Combine imputed 'Kms_Driven' values from df_encoded into df
df['Kms_Driven'] = df['Kms_Driven'].combine_first(df_encoded['Kms_Driven'])

# Calculate RMSE (Root Mean Squared Error)
rmse = np.sqrt(833765598.16)
print("RMSE:", rmse)
```
